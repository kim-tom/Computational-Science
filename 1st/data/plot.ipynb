```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.rcParams["font.size"] = 18
```

```python
ijk = pd.read_csv('ijk.tsv',header=None,sep='\t')
plt.plot(ijk[0][0:10],ijk[1][0:10],c='black',label='ijk',marker='^')
ikj = pd.read_csv('ikj.tsv',header=None,sep='\t')
plt.plot(ijk[0][0:10],ikj[1][0:10],c='black',label='ikj',marker='*')
kji = pd.read_csv('kji.tsv',header=None,sep='\t')
plt.plot(kji[0][0:10],kji[1][0:10],c='black',label='kji',marker='v')
kij = pd.read_csv('kij.tsv',header=None,sep='\t')
plt.plot(kij[0][0:10],kij[1][0:10],c='black',label='kij',marker='.')
jki = pd.read_csv('jki.tsv',header=None,sep='\t')
plt.plot(jki[0][0:10],jki[1][0:10],c='black',label='jki',marker='s')
jik = pd.read_csv('jik.tsv',header=None,sep='\t')
plt.plot(jik[0][0:10],jik[1][0:10],c='black',label='jik',marker='h')
plt.xlabel('N')
plt.ylabel('[ms]')
plt.legend()
plt.savefig('time.eps') #show
# ijk
```
